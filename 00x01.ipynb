#    un graphique à axes multiples (multi-axis chart)

### Données

```python
import pandas as pd 
import numpy as np
data = {
    "Mois": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Ventes (k€)": [20, 22, 25, 28, 50, 70, 80, 78, 60, 40, 30, 25],
    "Température (°C)": [5, 7, 10, 14, 18, 22, 25, 24, 20, 15, 10, 7],
    "Précipitations (mm)": [78, 60, 58, 55, 55, 40, 35, 45, 50, 60, 80, 85]
}
df = pd.DataFrame(data)
monti = np.arange(len(df["Mois"]))
monti 
```

```python
import matplotlib.pyplot as plt 
import plotly_express as px
```

###Premier axe : Ventes

```python
fig,ax1 = plt.subplots(figsize=(10,6))

ax1.bar(monti,df['Ventes (k€)'],color='green',label='Ventes (k€)')

ax1.set_xlabel("les mois")
ax1.set_ylabel("Ventes (k€)",color="green")
ax1.set_xticks(monti)
ax1.set_xticklabels(df["Mois"])
ax1.set_ylim(0,100)
plt.title("les vente par mois ")

plt.show()
```

### Deuxième axe : Température


```python
fig,ax1 = plt.subplots(figsize=(10,6))

ax1.bar(monti,df['Ventes (k€)'],color='green',label='Ventes (k€)')

ax1.set_xlabel("les mois")
ax1.set_ylabel("Ventes (k€)",color="green")
ax1.set_xticks(monti)
ax1.set_xticklabels(df["Mois"])
ax1.set_ylim(0,100)
ax2=ax1.twinx()
ax2.plot(monti,df['Température (°C)'],marker="o",color="red",label="Température (°C)")
ax2.set_ylim(0,30)
plt.title("les vente par mois ")

plt.show()
```

### utilisation fill_between

```python
plt.figure(figsize=(8,5))
plt.fill_between(df['Mois'],df['Précipitations (mm)'],color="#c6ebc5")
plt.title("Précipitations (mm) de chaque mois  ")
plt.ylabel("Précipitations (mm)")
plt.xlabel("les mois")
plt.show()
```

### Troisième axe : Précipitations

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Premier axe : barres pour les ventes
ax1.bar(monti, df['Ventes (k€)'], color='green', label='Ventes (k€)')
ax1.set_xlabel("Les mois")
ax1.set_ylabel("Ventes (k€)", color="green")
ax1.set_xticks(monti)
ax1.set_xticklabels(df["Mois"])
ax1.set_ylim(0, 100)

# Deuxième axe : ligne pour la température
ax2 = ax1.twinx()
ax2.plot(monti, df['Température (°C)'], marker="o", color="red", label="Température (°C)")
ax2.set_ylabel("Température (°C)", color="red")
ax2.set_ylim(0, 30)

# Troisième axe : remplissage pour les précipitations
ax3 = ax1.twinx()
ax3.fill_between(monti, df['Précipitations (mm)'], color="#c6ebc5", alpha=0.5, label='Précipitations (mm)')
ax3.spines["right"].set_position(("outward", 60))  # Déplace l'axe vers la droite
ax3.set_ylabel('Précipitations (mm)', color="#c6ebc5")
ax3.set_ylim(0, 100)

# Titre
plt.title("Les ventes par mois")

# Afficher le graphique
plt.show()
```
