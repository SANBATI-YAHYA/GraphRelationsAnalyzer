```python
import pandas as pd

data = {
    "Ville": ["Paris", "Lyon", "Marseille", "Lille", "Bordeaux", "Toulouse", "Nice"],
    "Latitude": [48.8566, 45.7640, 43.2965, 50.6292, 44.8378, 43.6047, 43.7102],
    "Longitude": [2.3522, 4.8357, 5.3698, 3.0573, -0.5792, 1.4442, 7.2620],
    "Taux de Criminalité": [650, 480, 700, 300, 450, 520, 680],
}

df = pd.DataFrame(data)
df
```

```python
import folium
```

```python
import folium
varf = folium.Map(location=[47,2],zoom_start=6)
for _,row in df.iterrows():
    
    folium.CircleMarker(location =[row["Latitude"],row["Longitude"]] , 
                        radius=row["Taux de Criminalité"]/100,
                        color="red",
                        fill=True,
                        fill_color = "red",
                        fill_opacity=1,
                        popup=f"{row["Ville"]}: {row["Taux de Criminalité"],"Taux de Criminalité"}"
                                  ).add_to(varf)
    
varf.save("map.html")
```

```python
coordinates = list(zip(df["Latitude"], df["Longitude"]))
folium.PolyLine(
    locations=coordinates,  
    color="red",
    
).add_to(varf)


varf.save("map_with_relations.html")
```
